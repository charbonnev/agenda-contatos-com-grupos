```python
import re

emails = 'etc@etc.com.br'

def verificar_email(emails):
    return re.search(r'[\w-]{1,20}@\w{2,20}\.\w{2,3}$', emails)

valido = verificar_email(emails)

if valido:
    print(emails, 'É um e-mail válido!')
else:
    print('Formato de e-mail inválido!:', emails)
```

```python
email1 = "charles.bonnevialle@gmail.com"
email2 = "charles@bol.com.br"
email3 = "conan_orcslicer@yahoo.com"

emails = [email1,email2,email3]

padrao = (r'[\w-]{1,20}@\w{2,20}\.\w{2,3}$')
validado = re.compile(padrao)

for e in emails:
  print('E-mail válido' if validado.search(e) else 'E-mail Inválido')
```
