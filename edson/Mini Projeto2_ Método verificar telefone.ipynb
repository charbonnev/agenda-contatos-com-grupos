```python
import re

telefones = "0800 606 5253"

def verificar_telefone(telefones):
    return re.search('^\(?[0-9]{0,4}\)? ?(?:[2-9]|9[1-9])[0-9]{3}\-?[0-9]{4}$', telefones)

correto = verificar_telefone(telefones)

if correto:
    
    print(telefones, 'É um telefone válido!')
else:
    print('Telefone inválido!:', telefones)
```

```python

telefones = [input(f"número {i}:") for i in range (int(input("Qtos números?")))]

padrao = '^\(?[0-9]{0,4}\)? ?(?:[2-9]|9[1-9])[0-9]{3}\-?[0-9]{4}$'
validador = re.compile(padrao)

for t in telefones:
  print('válido' if validador.search(t) else 'Inválido')
```
